```julia
using Sunny # The main package
using CairoMakie # Plotting package

units = Units(:meV, :angstrom)
```

```julia
cif = joinpath(@__DIR__, "Ho2Ti2O7_mp-33948_symmetrized.cif")
crystal = Crystal(cif; symprec=0.001)
Ho_sub = subcrystal(crystal, "Ho0")
```

```julia
crystal
```

```julia
dims = (2, 2, 2)  # Supercell dimensions
momentinfo = [1 => Moment(s=8, g=2)]  # Specify local moment information, note that all sites are symmetry equivalent
#sys_pyro = System(crystal, momentinfo, :dipole; dims)    # Make a system in dipole (Landau-Lifshitz) mode on pyrochlore lattice
sys_Ho = System(Ho_sub, momentinfo, :dipole; dims); # Same on MgCr2O4 crystal
```

```julia
print_symmetry_table(crystal, 5.9)
```

```julia
view_crystal(crystal)
```

```julia
J = +1.8*0.08617 # (meV)
set_exchange!(sys_Ho, J, Bond(2, 3, [0, 0, 0]))
view_crystal(sys_Ho)
```

```julia
randomize_spins!(sys_Ho)
minimize_energy!(sys_Ho)
plot_spins(sys_Ho; color=[S[3] for S in sys_Ho.dipoles])
```

## Langevin

```julia
langevin = Langevin(; damping=0.2, kT=16*units.K)
```

```julia
suggest_timestep(sys_Ho, langevin; tol=1e-2)
langevin.dt = 0.01676
```

```julia
energies = [energy_per_site(sys_Ho)]
for _ in 1:1000
    step!(sys_Ho, langevin)
    push!(energies, energy_per_site(sys_Ho))
end
```

```julia
lines(energies, color=:blue, figure=(size=(600,300),), axis=(xlabel="Timesteps", ylabel="Energy (meV)"))
```
